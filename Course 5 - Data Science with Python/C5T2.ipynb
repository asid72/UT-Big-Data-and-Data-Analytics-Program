```python
# Number of 1 -men and 2 - women
credit.groupby(['SEX'])['SEX'].count()
```

```python
# 1 Male and 2 Female
sns.factorplot('SEX', data=credit, kind='count', aspect=1.5)
```

```python
# How many are defaulted and not Defaulted for Next Month?
print(credit.groupby(['default payment next month'])['default payment next month'].count())
sns.factorplot('default payment next month', data=credit, kind='count', aspect=1.5)
# 6636 will be default and 23364 not default
```

```python
# Returns the only defaulted for next month
credit_defaulted = credit[credit['default payment next month']==1]
print("Total Defaulted: ",len(credit_defaulted))
```

```python
# Number of 1 -men and 2 - women that are defaulting?
print(credit_defaulted.groupby(['SEX'])['SEX'].count())
sns.factorplot('SEX', data=credit_defaulted, kind='count', aspect=1.5)
# Men 2873 and Female 3763
```

```python
#Defaulted by Education
print(credit.groupby(['default payment next month'])['EDUCATION'].count())
g = sns.factorplot('default payment next month',hue='EDUCATION', data=credit, kind='count', aspect=1.5)
g.set_xlabels('Class')
```

```python
#Defaulted by Education
print(credit_defaulted.groupby(['EDUCATION'])['EDUCATION'].count())
sns.factorplot('EDUCATION', data=credit_defaulted, kind='count', aspect=1.5)
```

```python
# break down of sex, education and marriage by the total count defaults
credittable = pd.crosstab(index=[
                                 credit_defaulted.MARRIAGE,credit.EDUCATION], columns=[credit_defaulted.SEX,pd.qcut(credit_defaulted['AGE'], q=4)])
credittable.unstack()
credittable
```

```python
# PSex and agegroups by default and non default
credittable2 = pd.crosstab(index=[credit['default payment next month']
                                 ], columns=[credit.SEX,pd.qcut(credit['AGE'], q=4)])
credittable2.unstack()
credittable2
```

```python
# Linear plot of age vs. default
sns.lmplot( 'default payment next month','AGE', data=credit)
```

```python
# Defaulted vs. Age grouped by Sex
sns.lmplot('AGE', 'default payment next month', data=credit, hue='SEX')
```
